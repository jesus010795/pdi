```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt

# /home/pigore/vision_python/PDI/03_histogram/img/2QV5YK~0.JPG
# img = cv.imread('./img/225V3H~L.JPG')
# img = cv.imread('./img/2QV5YK~0.JPG')
# img = cv.imread('./img/27P1TX~8.JPG')
img = cv.imread('../img/27P1TX~8.JPG')

img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
```

```python

clahe = cv.createCLAHE(clipLimit=2.0, tileGridSize=(5,5))
cl1 = clahe.apply(img)
res = np.hstack((img, cl1))


cv.imshow('Resultado', res)

cv.waitKey()
cv.destroyAllWindows()

his, bins = np.histogram(img.flatten(), 256, [0, 256])
plt.hist(img.flatten(), 256, [0, 256], color = 'b')
plt.xlim([0, 256])
plt.legend('H', loc = 'upper left')
plt.show()

his, bins = np.histogram(cl1.flatten(), 256, [0, 256])
plt.hist(cl1.flatten(), 256, [0, 256], color = 'b')
plt.xlim([0, 256])
plt.legend('H', loc = 'upper left')
plt.show()
```
