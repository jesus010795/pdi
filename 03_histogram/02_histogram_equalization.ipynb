# Ecualizacion de histograma

La ecualización del histograma es una técnica de procesamiento de imágenes que se utiliza para mejorar el contraste y la distribución de intensidad de píxeles en una imagen. Su objetivo principal es lograr un histograma más uniforme, lo que significa que se busca que cada nivel de intensidad tenga una cantidad similar de píxeles.

En este ejmplo tomaremos una imagen, la ecualizaremos y,luego mostraremos la imagen original junto a la ecualizada, ademas de mostrar los histogramas de ambas imagenes.

1. Ecualización y comparación visual (OpenCV):

    - Ecualizar histograma:

        - `eq = cv.equalizeHist(img)`: Esta línea ecualiza el histograma de la imagen cargada (img) usando la función cv2.equalizeHist de OpenCV y almacena el resultado en la variable eq.

    - Apilar imágenes horizontalmente:

        - `res = np.hstack((img, eq))`: Esta línea crea una nueva imagen (res) combinando horizontalmente la imagen original (img) y la imagen ecualizada (eq) utilizando la función np.hstack de NumPy.

    - Mostrar imágenes combinadas:

        - `cv.imshow('Resultado', res)`: Esta línea muestra la imagen combinada (res) en una ventana titulada "Resultado" usando la función cv2.imshow de OpenCV.

2. Cálculo y visualización de histogramas (Matplotlib):

    - Calcular histograma de la imagen original

        - `his, bins = np.histogram(img.flatten(), 256, [0, 256])`: Esta línea calcula el histograma de la imagen original (img) utilizando la función np.histogram de NumPy.

            - `img.flatten()`: Convierte la imagen a un array unidimensional.
            - `256`: Define el número de bins (contenedores) en el histograma (256 para representar todas las intensidades de píxeles posibles).
            - `[0, 256]`: Define el rango de valores de intensidad considerados (de 0 a 255).
        
        -  `plt.hist(img.flatten(), 256, [0, 256], color = 'b')`: Esta línea crea un diagrama de barras del histograma calculado usando `matplotlib.pyplot.`

            - Se utilizan los mismos parámetros que en el cálculo anterior.
            - `color = 'b'`: Define el color del diagrama como azul.

    - Ajustes del diagrama del histograma:

        - `plt.xlim([0, 256])`: Define el rango del eje X del diagrama de 0 a 256 para representar adecuadamente las intensidades.
        - `plt.legend('H', loc = 'upper left')`: Agrega una leyenda con la etiqueta "H" en la esquina superior izquierda.
        - `plt.show()`: Muestra el diagrama del histograma de la imagen original.
    
        - Este procedimiento tambien se realiza sobre la imagen ecualizada.

```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt

# /home/pigore/vision_python/PDI/03_histogram/img/2QV5YK~0.JPG
# img = cv.imread('./img/225V3H~L.JPG')
# /home/pigore/vision_python/PDI/03_histogram/img/27P1TX~8.JPG
# img = cv.imread('./img/27P1TX~8.JPG')
img = cv.imread('../img/27P1TX~A.JPG')


img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
```

```python
eq = cv.equalizeHist(img)
res = np.hstack((img, eq))

cv.imshow('Resultado', res)
# cv.imshow('Equalized Image', eq)

cv.waitKey()
cv.destroyAllWindows()

his, bins = np.histogram(img.flatten(), 256, [0, 256])
plt.hist(img.flatten(), 256, [0, 256], color = 'b')
plt.xlim([0, 256])
plt.legend('H', loc = 'upper left')
plt.show()

his, bins = np.histogram(eq.flatten(), 256, [0, 256])
plt.hist(eq.flatten(), 256, [0, 256], color = 'b')
plt.xlim([0, 256])
plt.legend('H', loc = 'upper left')
plt.show()
```

Este ejemplo demuestra cómo ecualizar el histograma de una imagen en OpenCV y luego comparar visualmente la imagen original con la ecualizada. Además, utiliza matplotlib para calcular y mostrar los histogramas de ambas imágenes, lo que permite analizar la distribución de intensidades de píxeles antes y después de la ecualización.

Esto puede ser útil en diversas situaciones:

1. **Mejorar el contraste**:

En imágenes con bajo contraste, la ecualización del histograma puede extender el rango de intensidades de píxeles, haciendo que las áreas claras y oscuras sean más distintivas y mejorando la percepción visual general.

2. **Resaltar detalles:**

Al distribuir las intensidades de manera más uniforme, la ecualización del histograma puede revelar detalles que antes no se veían claramente debido a la falta de contraste o a la concentración de píxeles en ciertos niveles de intensidad.

3. **Compensar la iluminación**:

En imágenes con iluminación no uniforme, la ecualización del histograma puede ayudar a compensar las variaciones de iluminación y mejorar la consistencia de la imagen.

4. **Preparación para otras técnicas**:

La ecualización del histograma puede ser un paso previo útil para otras técnicas de procesamiento de imágenes, como la segmentación de imágenes o el análisis de texturas, ya que proporciona una base más uniforme para trabajar.

- **Ejemplos de aplicaciones**:

- Imágenes médicas: Para mejorar la visibilidad de estructuras anatómicas en radiografías o resonancias magnéticas.
- Fotografías: Para realzar detalles en paisajes, retratos o fotografías con poca luz.
- Escaneo de documentos: Para mejorar la legibilidad del texto en documentos antiguos o desgastados.
